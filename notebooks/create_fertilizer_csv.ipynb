```python
import pandas as pd
import numpy as np
from math import ceil
```

```python
crop = pd.read_csv('/content/crop_recommendation.csv')
crop.head()
```

```python
crops = list(crop.label.unique())
crops
```

```python
def calcMean(crop_name):
	N = crop[crop['label'] == crop_name]['N']
	P = crop[crop['label'] == crop_name]['P']
	K = crop[crop['label'] == crop_name]['K']
	n = N.size
	meanN = ceil(sum(N)/n)
	meanP = ceil(sum(P)/n)
	meanK = ceil(sum(K)/n)
	return (meanN, meanP, meanK)
```

```python
my_dict = {}
my_dict['Crop'] = []
my_dict['N'] = []
my_dict['P'] = []
my_dict['K'] = []
for crp in crops:
	(n,p,k)= calcMean(crp)
	my_dict['Crop'].append(crp)
	my_dict['N'].append(n)
	my_dict['P'].append(p)
	my_dict['K'].append(k)


df = pd.DataFrame.from_dict(my_dict)
df.head()
```

```python
df.to_csv('/content/my_fertilizers.csv', index=False)
```
